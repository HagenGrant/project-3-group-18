```python
import pandas as pd
import numpy as np
```

```python
file = "tornados.csv"
df = pd.read_csv(file, encoding="cp1252")
df.head()
```

```python
#Dropping Columns
df = df.drop(columns=['dy', 'tz', 'datetime_utc', 'stf', 'time', 'f1', 'f2', 'f3', 'f4', 'fc', 'ns', 'sn'])
df.head()
```

```python
#Renaming columns
df = df.rename(columns = {"om" : "tornado_id", "mo": "month", "st": "state", "mag": "category", "inj": "injuries", "fat": "fatalities", 
                          "slat": "start_lat", "slon": "start_longitude", "elat": "end_latitude", "elon": "end_longitude", "len": "distance_traveled", 
                          "wid": "width"})
df.head()
```

```python
#Turning NULL losses and  into 0
df['loss'] = df['loss'].fillna(0)
df['category'] = df['category'].fillna(0)
df.head()
```

```python
df.info()
```

```python
# df = df.dropna(axis='columns')
df.info()
```

```python
#Chaning data types
df['date'] = pd.to_datetime(df['date'])
```

```python
#Creating dictionary for season.
north_seasons = {'spring': [3, 4, 5],
                 'summer': [6, 7, 8],
                 'fall': [9, 10, 11],
                 'winter': [12, 1, 2]}
```

```python
#Running through each row to apply season with the dictionary created previously.
seasonz = []
for row in df.index:
    if df.at[row, 'month'] in north_seasons['spring']:
        seasonz.append('spring')
    elif df.at[row, 'month'] in north_seasons['summer']:
        seasonz.append('summer')
    elif df.at[row, 'month'] in north_seasons['fall']:
        seasonz.append('fall')
    elif df.at[row, 'month'] in north_seasons['winter']:
        seasonz.append('winter')
    else:
        seasonz.append('Unknown')
df['seasons'] = seasonz
```

```python
df.head()
```

```python
df = df.loc[df["yr"] > 2015, :]
df.info()
```

```python
#Exporting the clean dataset to a new csv file
df.to_csv("tornadoes_clean.csv", index_label="id")
```
